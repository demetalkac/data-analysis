```python
# Bu tam olmadı ama silmek istemedim :)
m = np.linspace(0, 10, 11) 
c = 3 * 10**8             
E = m * c**2 


# Figure oluştur ve boyutunu ayarla
plt.figure(figsize=(6, 4))

# Semilog çizim, y ekseni için logaritmik ölçek =>chatgpt den burası
plt.semilogy(m, E, color='green', linewidth=3)

# Eksen etiketleri
plt.xlabel("M in g")
plt.ylabel("E in J")

# Başlık
plt.title("E=mc^2")


plt.grid(True, which="both", linestyle='-', linewidth=0.5)

plt.show()
```

```python
m = np.linspace(0, 10, 11)  # 0'dan 10'a kadar eşit aralıklı 11 sayı
c = 3 * 10**8               # Işık hızı (m/s)
E = m * c**2                # Enerji hesaplaması

# Grafik oluşturma
plt.plot(m, E, color="green", lw=3)  # Enerji-madde grafiği
plt.title("E = mc^2")                # Grafik başlığı
plt.xlabel("M in g")                 # x-ekseni etiketi
plt.ylabel("E in J")                 # y-ekseni etiketi
plt.xlim(0, 10)                      # x-ekseni sınırları
plt.yscale("log")                    # y-eksenini logaritmik ölçeğe ayarlama
plt.grid(which="both", axis="y")     # y-ekseninde ızgara çizgilerini ayarlama
plt.show()
```

## Aşağıdaki değişkenlere göre ilgili platlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
%matplotlib inline
plt.figure(figsize = (6, 4))

plt.plot(labels, july16_2007, "c",label="july16_2007" , lw=1)
plt.plot(labels, july16_2020, color='orange', label="july16_2020", lw=1)

plt.legend(loc = "right")
plt.show()
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python

fig, ax =plt.subplots(2, 1) 

ax[0].plot(labels, july16_2007, "b")
ax[0].set_title("july16_2007")


ax[1].plot(labels, july16_2020, "c")
ax[1].set_title("july16_2020")




plt.tight_layout()
plt.show()
```

```python
#olması gereken
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python

fig, ax1 = plt.subplots()

# İlk çizgiyi çiz
ax1.plot(labels, july16_2007, "b", label="July 16, 2007")
ax1.set_xlabel("Time ")
ax1.set_ylabel("July 16 2007", color="b")
ax1.tick_params(axis="y", labelcolor="b")

# İkinci yatay eksenli alt grafiği oluştur
ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, "r", label="July 16, 2020")
ax2.set_ylabel("July 16 2020", color="r")
ax2.tick_params(axis="y", labelcolor="r")

# Başlık
#plt.title("July 16, 2007 vs July 16, 2020")

plt.show()
```

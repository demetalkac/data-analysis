# Functional Method

```python
# Functional method kullanırken, subplot() methodunu cagırıp icine olusturacagım 
# grafigin kac satırda  ve kac sütun da göstermek istedigimi belirtiyorum. Eger birden fazla ise cizmek 
# istedigim grafikler son parametreye de kacıncı sırada olması gerektigini yazıyorum.

plt.subplot(2, 1, 1) # subplot(): icine satır ve sütun sayısını belirtiyoruz.
plt.plot(age, salary, "r")  # subplot(2, 1, 1) : grafikleri 2 satırda  ve 1 sütunda göster diyoruz. sondaki 1 ise 1. grafigim bu olsun diye belirtmis oluyoruz.

plt.subplot(2, 1, 2)  # subplot(2, 1, 2) : grafikleri 2 satırda  ve 1 sütunda göster diyoruz. sondaki 2 ise 2. grafigim bu olsun diye belirtmis oluyoruz.
plt.plot(age, salary_2, "b")

plt.show()
```

# Object Oriented Method (OOM)

```python
# OOM model ile daha az kod yazmıs oluyoruz.
# OOM modelde, fig, ax imi önce tanımlıyorum. 
# subplots() ile kac sub plottan olusacagını belirliiyorum.
# ax larımı yazıyorum.

fig, ax =plt.subplots(2, 1) # subplots(2, 1) : 2 satır, 1 sütundan olusan

ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "b")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2)

ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "b")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,3) # 2 satır, 3 sütun

ax[0][0].plot(age, salary, "r")
ax[0][1].plot(age, salary_2, "b")
ax[0][2].plot(age, salary, "r")

ax[1][0].plot(age, salary_2, "b")
ax[1][1].plot(age, salary, "r")
ax[1][2].plot(age, salary_2, "b")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(3,2)  #3 satır , 2 sütun

ax[0][0].plot(age, salary, "r")
ax[0][1].plot(age, salary_2, "b")

ax[1][0].plot(age, salary_2, "b")
ax[1][1].plot(age, salary, "r")

ax[2][0].plot(age, salary_2, "b")
ax[2][1].plot(age, salary, "r")

plt.tight_layout()
plt.show()
```

```python
x = np.arange(0, 30, 0.2)

fig, ax = plt.subplots(figsize = (14, 6)) # burda subplots() icinde boyutları belirtmedik(örn:subplots(3,2) gibi),
     # o nedenle tek grafikte verdi bütün karsılastırmak istedigimiz verileri
     # Bu yüntemi 2 veriyi, mesela 4 kisinin maaslarını aynı grafikte karsılastırmak istersek kullanabiliriz.                                  


ax.plot(x, x-1, color ="green", lw = 3, ls = "-")  
ax.plot(x, x-2, color ="red", lw = 3, ls = "-.")
ax.plot(x, x-3, color ="blue", lw = 3, ls = ":")
ax.plot(x, x-4, color ="purple", lw = 3, ls = "--");

# ax.plot(x, x-1, : normalde x, x'si verirken, x-1 ise y koordinatını veriyor.
# Fakat burda x, x eksenini verirken, x-1 cizgi grafik ile cizdigimiz cizgiler arasında bosluk olsun diye ekledi hoca
# Eger bir eksen varsa zaten (burdaki gibi) y eksenini belirtmesek ,o y eksenine de x eksenine verdigimizi kullanır
```

# Multible Plots

```python
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)

fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize= (20, 20))

ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart
ax1[0, 0].set_title("Square Function") # setting title of subplot
ax1[0, 0].set_xlabel("X", fontsize = 18)
ax1[0, 0].set_ylabel("Y", fontsize = 18)

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart
ax1[0, 1].set_title("Cubic Function") # setting title of subplot
ax1[0, 1].set_xlabel("X", fontsize = 18)
ax1[0, 1].set_ylabel("Y", fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart
ax1[0, 2].set_title("Sine Function") # setting title of subplot
ax1[0, 2].set_xlabel("X", fontsize = 18)
ax1[0, 2].set_ylabel("Y", fontsize = 18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart
ax1[1, 0].set_title("Cosine Function") # setting title of subplot
ax1[1, 0].set_xlabel("X", fontsize = 18)
ax1[1, 0].set_ylabel("Y", fontsize = 18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart
ax1[1, 1].set_title("Tangent Function") # setting title of subplot
ax1[1, 1].set_xlabel("X", fontsize = 18)
ax1[1, 1].set_ylabel("Y", fontsize = 18)


ax1[1, 2].plot(x, y6, "g-") # set the color of line chart
ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot
ax1[1, 2].set_xlabel("X", fontsize = 18)
ax1[1, 2].set_ylabel("Y", fontsize = 18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart
ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot
ax1[2, 0].set_xlabel("X", fontsize = 18)
ax1[2, 0].set_ylabel("Y", fontsize = 18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart
ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot
ax1[2, 1].set_xlabel("X", fontsize = 18)
ax1[2, 1].set_ylabel("Y", fontsize = 18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart
ax1[2, 2].set_title("Exp Function") # setting title of subplot
ax1[2, 2].set_xlabel("X", fontsize = 18)
ax1[2, 2].set_ylabel("Y", fontsize = 18)
plt.show()





```

```python
fig, ax = plt.subplots()

ax.scatter(tips.total_bill, tips.tip) # plot umu scatter bu verdiklerime göre cizecek

x=tips.total_bill # formülu uygulayacagı x ve y yi belirledik
y=tips.tip

m, b = np.polyfit(x, y, 1) #Basit doğrusal regresyon hesaplaması
# m :egim-slope
# b :intercept parametresi,x ve y nin kesisim noktaları
# Regresyon Formülünün Meali: Bagimli Degisken = Kesim + Egim X Bagimsiz degisken
ax.plot(x, m*x + b, c="r")
plt.show()


# nampay kütüphanesindeki polyfit() methodu ile linear bir cizgi ile verilerin yogun gectigi yeri cizebiliriz.
#linear regrassion formulu(basit regrasion formulü olarakta bilinir):Y=β0+β1X+ε  =>bunu ml de ögrenecekmisiz.
```

```python
# 4 meyveyi pasta grafik seklinde olusturma: pie()

x= np.array([100, 100, 100, 100]) # bu ise herbir dilim icin 100 mlimlik dizi olusturacak. Burda esit dilimler olarak ayarladık
mylabels = ["Apples","Bananas","Cherries", "Melon"] # pasta dilimlerini etiketlemek icin kullandık meyve isimlerini

plt.figure(figsize=(10,8))

plt.pie(x, labels= mylabels, labeldistance= 0.5, autopct= "%.2f") 
# labeldistance= 0.5 => label(mylabels olarak verdiklerimiz) ile rakamsaal veri arasındaki uzaklık, label ların merkeze olan uzaklıgını ayarlamak icin kullnyrz.
# autopct= "%.2f" => virgülden sonraki rakaam sayısı, burda % de 2 

plt.show()
```

```python
x = np.array([100, 100, 100, 100])
mylabels = ["Apples", "Bananas", "Cherries", "Melon"]

plt.figure(figsize=(10, 8))

myexplode = [0.8, 0.3, 0, 0.5] # =>dilimleri ayıma. sag ustten basliyor, sonra saat yonun tersi diye anladim
# explode u float seklinde olusturuyoruz.


plt.pie(x, labels = mylabels, labeldistance = 0.5, explode = myexplode, autopct = "%.2f")

#explode : dilimleri ayırdı
plt.show()
```

```python
# pie grafigi renklendirdik legant kullanarak:

plt.figure(figsize=(10, 8))

color_list = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode_list = [0.5,0,0,0]


y = np.array([25, 35, 25, 15])

mylabels = ["Apples", "Bananas", "Cherries", "Melon"] # mylabels : üzerinde yazacak yazılar


plt.pie(y, 
        labels = mylabels, 
        labeldistance = 0.5,
       autopct = '%.1f',
       startangle = 90,
       shadow = True,    # gölgelendirme yapıyoruz, Treu secerek.
       pctdistance = 0.3,
       colors = color_list,
       explode=explode_list)


plt.legend(loc = (1))

plt.show()
```

## <font color='blue'> <b>linspace() </b><font color='black'>

Bu fonksiyon, özellikle grafik çizme ve veri analizi gibi durumlarda yararlıdır.



```python
np.linspace(1,20)
```

```python
np.linspace(1,20, 5)
```

```python
np.linspace(1,20, 5 , dtype=int)    # dtype="i" olarak da yazılabilir
```

```python
int(1.8)
```

```python
int(0.8)
```

```python
np.linspace(1,20, 5, retstep=False)
```

## <font color='blue'> <b>logspace() </b><font color='black'>
logspace() logaritmik aralıklarda bir dizi oluşturur. 

logspace() fonksiyonu, linspace() fonksiyonuna benzer şekilde çalışır, ancak aralıklar logaritmik olarak hesaplanır.

Logspace : Başlangıç değeri 10a ,bitiş değeri değeri 10b olan n elemanlı ve elemanları arasındaki katları eşit olan bir dizi oluşturur.


logspace() => daha cok grafik cizmede, büyük datalar ile kullanılır
-defoult degeri onluk tabanında dır, ama istersek tabanı degistirebiliriz.

```python
np.logspace(2,3) # 10^2 ile 10^3 arasında 50 adet logaritmik aralıklı deger olusturuyor
```

```python
np.logspace(2,3, num=10, base=2)# => 2 üzeri 2=4 oldugu icin 4'den basladı, ve 2 üzeri 3 =>8 oldugu icinde 8'e kadar deger aralıgı verdi.
                                #    Ve 10 deger diye belirttigimiz icinde 4 ve 8 arasında 10 deger verdi.
                                # logaritmik tabanı degistirebiliriz.
                                #  num=10 => 10 tane olustur
                                # base=2. =>base matematiksel taban anlamında
                                # esit aralıga göre deger üretir
```

## <font color='blue'> <b>eye() </b><font color='black'>
- Bu fonksiyon, belirtilen boyutta bir birim matris (identity matrix) oluşturur. 
- Birim matriste, ana köşegen üzerindeki elemanlar 1'dir ve diğer tüm elemanlar 0'dır.


```python
np.eye(3) # eye() köselerinden capraz olarak 1'ler koyarak, ve diger kalan yerleri 0'lar ile doldurarak bir array olusturur
```

```python
np.eye(3,3) # eye(3) = eye(3,3) 
```

```python
np.eye(3, k=1) # k=1 => 0.indexten degil 1. indexten baslayarak 1'leri yerlestirir.
               # k=2 => 2. indexten baslayarak 1'leri yerlestirir.
```

```python
np.eye(3, k=-1) # k=-1 => -1 ise ,0. degil 1. satırdan basla 1'leri yazmaya demek oluyor. 
                #         -2 desek 2. satırdan basla demis oluruz.
```

## <font color='blue'> <b>full() & full_like() </b><font color='black'>

**full_like():** Verilen bir dizinin şekli ve türüne sahip, ancak belirtilen değerle doldurulmuş bir dizi oluşturur.


```python
np.full((2,2)) # array'i hangi degerlerle olusturacagımızı söylemedigimiz icin hata verir
               # full(()) array i kendi istedigimiz bir deger ile dolduran bir 
               # fully baska sekillerde de olusturulabilir, "inf" fonksiyonunu kullanarak infinitive ile doldur, diyebiliriz
               # Veya icine sadece bir sayı degil 2 deger vererekde doldurmasını isteyebiliriz: np.full((2,2), [1,2]) yani [1,2]'ler ile olustur.'
```

```python
 np.full((2,2), [1,2]) # Veya icine sadece bir sayı degil 2 deger vererekde doldurmasını isteyebiliriz: 
                       # np.full((2,2), [1,2]) yani [1,2]'ler ile olustur.'
```

```python
 np.full((2,2), np.inf) # "inf" ler ile doldur demis oluyoruz.
```

## <font color='blue'> <b>itemsize</b><font color='black'>
    
itemsize(), NumPy dizisinin her bir elemanının bellekte kaç byte kapladığını gösterir. 

Bu özellik, NumPy dizisinin veri tipine (dtype) bağlıdır ve bu veri tipinin boyutuna göre değişir.

itemsize(), bir NumPy dizisinin bellekte kapladığı toplam alanı hesaplamak için de kullanılabilir. Bu hesaplama, dizinin eleman sayısının itemsize() ile çarpımı ile yapılabilir.

float32 tipi 32 bit (4 byte) bellek kullanır ve yaklaşık 7 basamak hassasiyetle ondalık sayıları temsil edebilir. 

32 bitlik bir sayı, 2^32 farklı sayıyı ifade edebilir.

float64 tipi ise 64 bit (8 byte) bellek kullanır ve yaklaşık 15-16 basamak hassasiyetle ondalık sayıları temsil edebilir

64 bitlik bir sayı, 2^64 farklı sayıyı ifade edebilir.

```python
arr
```

```python
arr.itemsize # arr icindeki tek bir elemanın kapladıgı yeri(byte) verir
```

```python
arr.dtype
```

```python
arr.size
```

```python
arr.size * arr.itemsize # itemsize bir degerin kapladıgı yeri verir, o nedenle size ile itemsize ı carparsak 
                        # tüm arrayin kapladıgı alanı buluruz.
```

# <font color='green'> <b>Concatenation of the Arrays</b><font color='black'>
concatenate() fonksiyonu, iki veya daha fazla diziyi birleştirmek için kullanılır. 

Fonksiyonun ilk parametresi birleştirilecek dizilerin bir listesidir. 

İkinci parametre ise, birleştirme işleminin hangi eksen veya boyuta göre yapılacağını belirten bir parametredir.

Örneğin, axis=0 parametresi, dizilerin satır bazında (dikey birleştirme) birleştirilmesini, axis=1 parametresi ise sütun bazında (yatay birleştirme) birleştirilmesini sağlar.



 axis=0 parametresi satır bazında birlestirme yapar, defoult degeri 0'dır yazmasak da olur. Satır olarak ekler
 axis=1 sutun olarak yana ekleme yapar.

```python
x = np.array([1,2,3]) # x tek boyutlu array
x
```

```python
y = np.array([4,5,6]) # y tek boyutlu array
y
```

```python
np.concatenate(x,y) 
```

```python
xy = np.concatenate((x,y)) # tuble veya list icinde vermek gerekiyor
xy                         # tek boyutlu arrayleri birlestirdik
```

```python
xy = np.concatenate((x,y), axis=1) # tek boyutlu diziler icin sadece axis=0 kullanılır. 
                                   # Cünkü satır ve sutunları yok tek boyutlu array de
                                   # axis=0 i yazmaya gerek yok direk oyle calısır
```

```python
a = np.array([[1,2], [3,4]])
a
```

```python
b = np.array([[5,6], [7,8]])
b
```

```python
np.concatenate((a,b))# axis yazmadık burda, defoult olarak axis=0 dır. Yani alt alta ,satır olarak ekler.
```

```python
np.concatenate((a,b),axis=1) # axis=1 ise yan yana, sutun olarak ekler
```

```python
np.concatenate((a,b),axis=None) # axis= None => 2 boyutlu array i tek boyutlu bir arraye dönüstürür.
```

# <font color='green'> <b>Splitting of the Arrays</b><font color='black'>
[numpy.split 01](https://www.w3resource.com/numpy/manipulation/split.php),
[numpy.split 02](https://sparkbyexamples.com/numpy/numpy-split-array/),
[numpy.split 03](https://numpy.org/doc/stable/reference/generated/numpy.array_split.html),
[numpy.split 04](https://www.w3schools.com/python/numpy/numpy_array_split.asp)

**split():**, numpy dizisini belirtilen eksende eşit parçalara bölmek için kullanılır.

**array_split():** fonksiyonu, NumPy'da bir diziyi belirtilen sayıda eşit olmayan parçalara bölmek için kullanılır. Bu fonksiyon, diğer bölme fonksiyonlarından (hsplit, vsplit gibi) farklı olarak belirli bir eksende değil, kullanıcının belirttiği sayıda parçaya böler.

***vsplit():** NumPy'da bir diziyi yatay olarak (yani satırlar üzerinden) birden çok parçaya ayırmak için kullanılır.

***hsplit():** NumPy'da bir diziyi dikey olarak (yani sütunlar üzerinden) birden çok parçaya ayırmak için kullanılır. Temel olarak, bir matrisi sütunlarını kullanarak belirtilen indekslerde böler.


   

```python
a = np.array([1,2,3,4,5,6,7,8,])
a
```

```python
np.split(a,4) # split() i arrayi bölmek istedigimizde esit parcalara ayrılabiliyorsa kullanırız
```

```python
np.split(a,3)# split esit parcalara bölemedigi icin hata verdi
```

```python
np.array_split(a,3) # esit bölünemeyen arrayleri gene de bölmek istersek array_split()kullanırız
```

```python
x = np.arange(20).reshape(5,4)
x
```

```python
np.split(x, [1,3])# x array ile calıs: 0'dan basla 1. indexe kadar olanı al, 
                  # 1.indexten 3. indexe kadar olanı ayrı arra yap , ve 3. arrayden de kalan kısmı al demis olduk
```

```python
d = np.arange(16.0).reshape(4, 4)
```

```python
np.vsplit(d,2)
```

```python
np.hsplit(d,2)
```
